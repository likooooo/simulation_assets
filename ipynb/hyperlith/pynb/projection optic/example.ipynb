- NA  
The numerical aperture (NA) (the sin of the angle of the most oblique plane wave captured by the optical system).  This is the image-side NA (usually specified by stepper/scanner users).  The object-side NA (usually specified for microscopy/inspection users) is related as:  image_side_NA=demagnification * object_side_NA.   For an ArF lithography stepper/scanner, it is typically in the range from 0.7 to 1.35.  For immersion lithography the NA can be greater than 1.0 and must be less than or equal to the index of the superstrate material specified on the Thin Film Stack at Wafer Tab.  For a magnifying system, Image Demagnification is less than 1.0.  Also, for a microscope/inspection system, one typically specifies NA at the object (not the image) and so the user must be careful to remember that the program is expecting the image-side NA.  For example:  a 100X magnifying system with an object-side NA of 0.75 would have Image Demagnification set to 0.01 and NA set to 0.75*0.01 which equals 0.0075.

- low-order blocking aperture  
for dark field imaging one would change this to a value greater than 0.0.

- Image Plane Defocus  
The deviation from the ideal image plane.  The quantity is specified in the same units as was used to specify the wavelength in the Mask/Object Main Tab.  For a stepper/scanner, an ideal binary mask (i.e. thin mask or Kirchhoff mask) with easily resolvable features would have best focus (i.e. image would have maximum contrast) at the ideal image-side focal plane (specified by the lens law).  Positive defocus corresponds to the wafer moving closer to the lens.  Said another way, positive defocus corresponds to the waist of the image moving down through the resist towards the substrate. 

- Image Demagnification  
The projection optic's demagnification.  See NA above.  A 4X stepper (a reduction optical system) has a demagnification of 4.0. 

- Use Radiometric Correction  
Account for conservation of energy as plane waves change direction as they pass through the projection optic.  This should be left checked. 

- EUV axis theta (degrees)/EUV axis phi (degrees)  
It is only necessary to specify these values when simulating off-axis EUV systems where Non-Constant Scattering Coefficients are used.  In this case, there are multiple angles of incidence that are simulated in TEMPESTpr2 (a TE and a TM plane wave at each angle).  These two fields specify the angle of the central ray so that the appropriate set of scattering coefficients is chosen for each particular source integration point.  Typically they would be set to theta=6 degrees and phi=0 degrees.  See EUV with Non-Constant-Scattering-Coefficients example.
