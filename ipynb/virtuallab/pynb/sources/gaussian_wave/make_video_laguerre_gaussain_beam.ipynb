```python
%matplotlib inline
from simulation_toykit_sources import *
sys.path.append('../virtuallab/tools')
from common import *
import sys


degree       = np.pi/180
wavelength   = 0.5
k      = 2 * np.pi/wavelength
start  = [-5, -5] # 'um'
end    = [5, 5]
shape  = [1000, 1000]

step   = [(b-a)/N for a,b,N in zip(start, end, shape)]
meta   = {
    "dx" : step[0] * 1e-3, # set display unit to 'mm'
    "dy" : step[1] * 1e-3,
    "nx" : shape[0],
    "ny" : shape[1]
}

w0 = 2
p, l = 6, 7
def update_frame(p, l, z):
    U = laguerre_gaussian_wave.generate(p, l, wavelength, z, w0 ,start, end, shape)
    U = np.array(list(U)).reshape(*shape)
    U /= np.max(np.abs(U))
    return np.real(U)
```

```python
from make_video import make_video

N = 300
video_path = f"laguerre_gaussain_beam.mp4" 

M2 = gaussian_wave.LG_M2(p, l)
rz = gaussian_wave.rz(k, w0, M2)
sweep_data = []
for i in range(N):
    sweep_data.append({
        "p": p,
        "l": l,
        "z": i * rz / N
    })

make_video(
    output_filename=video_path,
    frame_callback=update_frame,
    sweep_table=sweep_data,
    frame_duration=10/N
)
```
