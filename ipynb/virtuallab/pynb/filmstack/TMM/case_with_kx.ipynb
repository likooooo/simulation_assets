```python
import numpy as np
from simulation import *
from math import *
meterial = meterial_s
degree = pi/180

def print_result(rs, ts, rp, tp):
    print("  \t        s    \t                p")
    print("r \t{:.5f}\t{:.5f}".format(rs, rp))
    print("t \t{:.5f}\t{:.5f}".format(ts, tp))
```

```python
def fresnel_equaltion_example(th_0 = np.arcsin(0.24), check_only = False):
    air , ResistSL  = meterial(), meterial()
    air.nk     = 1
    ResistSL.nk= 0.98 + 0.005j
    
    meterials = [air, ResistSL]
    dir = TMM_propagate_direction(meterials, th_0)
    wl = 13.5
    tmm_list1 = TMM_interface_transfer_matrix_with_thickness_without_multiply_s(meterials, dir, wl)
    tmm_list = TMM_apply_multilayer_effect(tmm_list1)
    rs, ts = TMM_get_r_t_from_tmm(tmm_list[-1])
    if check_only: print(f"rs={rs}\tts={ts}\tcheck={ts-rs} == 1")

    tmm_list1 = TMM_interface_transfer_matrix_with_thickness_without_multiply_p(meterials, dir, wl)
    tmm_list = TMM_apply_multilayer_effect(tmm_list1)
    rp, tp = TMM_get_r_t_from_tmm(tmm_list[-1])
    if check_only: print(f"rp={rp}\ttp={tp}\tcheck={meterials[-1].nk/meterials[0].nk*tp - rp} == 1")
    else :         print_result(rs, ts, rp, tp)

for i in range(10):
    fresnel_equaltion_example(np.arcsin(0.099*i), True)

print("\n\n-----------------")
fresnel_equaltion_example(np.arcsin(0.24))
```

```python
def reflection_example():
    air , ResistSL, BARC, C, Si,  = meterial(), meterial(), meterial(), meterial(), meterial()

    air.nk     = 1
    ResistSL.nk= 0.98 + 0.005j
    BARC.nk    = 1.2  + 0.05j
    C.nk       = 0.962+ 0.007j
    Si.nk      = 0.999+ 0.002j

    ResistSL.depth = 40
    BARC.depth     = 20 
    C.depth        = 30 

    
    th_0 = np.arcsin(0.24)
    meterials = [air, ResistSL, BARC, C, Si]
    dir = TMM_propagate_direction(meterials, th_0)
    wl = 13.5
    tmm_list1 = TMM_interface_transfer_matrix_with_thickness_without_multiply_s(meterials, dir, wl)
    tmm_list = TMM_apply_multilayer_effect(tmm_list1)
    rs, ts = TMM_get_r_t_from_tmm(tmm_list[-1])

    tmm_list1 = TMM_interface_transfer_matrix_with_thickness_without_multiply_p(meterials, dir, wl)
    tmm_list = TMM_apply_multilayer_effect(tmm_list1)
    rp, tp = TMM_get_r_t_from_tmm(tmm_list[-1])
    print_result(rs, ts, rp, tp)

    print("-------------------------\nresult from kx's matlab:")
    print('''
        K>> r_s

        r_s =

        0.082827018253689 + 0.095067957492916i

        K>> r_p

        r_p =

        0.074057753110956 + 0.086317925914788i

        K>> t_s

        t_s =

        -0.351789179415013 - 0.358934290265760i

        K>> t_p

        t_p =

        -0.352694213669353 - 0.360272632413191i
''')
reflection_example()
```
