```python
degree_to_rad = pi / 180
def fourier_shift(theta, phi):
    return sin(theta) * cos(phi), sin(theta) * sin(phi)

wavelength = 13.0
theta      = 6 
phi        = 90
NA         = 0.9
M          = 4
k0         = 2 * pi / wavelength # 角度的改变对应频谱的移动, 与 NA 无关!
shiftx, shifty = [n * k0 for n in fourier_shift(theta * degree_to_rad, phi * degree_to_rad)]

print(f"* chief-ray test.\n    chief-ray is ({theta}, {phi})\n    wavelength is {wavelength}\n    demangnification is {M}\n    NA is {NA}\n    shift in mask P.O.V is ({shiftx}, {shifty})")
```

```python


def mask_pov_shift_to_sigma_xy(shiftx, shifty):
    mask_na = NA / M
    k0 = 2 * pi / wavelength * mask_na
    return shiftx / k0, shifty / k0 

delta_sigma_x, delta_sigma_y = mask_pov_shift_to_sigma_xy(shiftx, shifty)
print(f"* chief-ray test.\n    chief-ray in Wafer P.O.V is ({delta_sigma_x}, {delta_sigma_y})") 
```
